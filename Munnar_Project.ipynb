```python
import pandas as pd
import numpy as np
import sys
```

```python
df = pd.read_csv('/content/Data Analyst munnar Dataset.csv')
df
```

```python
df.rename(columns = {'Hotel Name\t\t\t\t\t\t\t\t\t' : 'Hotel Name'},  inplace = True)
df
```

```python
print(type(df))
print(sys.getsizeof(df))
print(df.columns)
print(df.shape)
```

```python
from timeit import default_timer as timer
start = timer()
df['Star Rating'].isna().sum()
print(timer() - start)
```

```python
df['Tax'].isna().sum()
```

```python
df.reset_index(inplace = True, drop = True)
df
```

```python
df.nunique()
```

```python
df['Tax'] = df['Tax'].apply(lambda x : str(x))
df['Tax'] = df['Tax'].apply(lambda x : x.replace(",", ""))
df['Tax'] = df['Tax'].astype('float64')
df['Tax']
```

```python
df['Price'] = df['Price'].apply(lambda x : str(x))
df['Price'] = df['Price'].apply(lambda x : x.replace(",", ""))
df['Price'] = df['Price'].astype('float64')
df['Price']
```

```python
def m_into_km(x):
  if x.split()[1] == 'm':
    return float(x.split()[0])/1000
  elif x.split()[1] == 'km':
    return x.split()[0]

df['Distance to Landmark'] = df['Distance to Landmark'].apply(m_into_km)
df['Distance to Landmark'] = df['Distance to Landmark'].astype('float64')
df['Distance to Landmark']
```

```python
df.info()
```

```python
df['Star Rating'].fillna(df['Star Rating'].value_counts().index[0], inplace = True)
df.info()
```

```python
df['Tax'].fillna(df['Tax'].median(), inplace = True)
df.info()
```

```python
sys.getsizeof(df)
```

```python
def color_map(x, avg):
  if x < avg:
    return 'color: ' + 'red'
  else:
    return 'color: ' + 'green'

avg_val = df['Tax'].mean()
```

```python
rating_desc = df['Rating Description'].unique()
rating_desc
```

```python
def rating_desc_to_num(x):
  if x == 'Average':
    return 1
  elif x == 'Good':
    return 2
  elif x == 'Very Good':
    return 3
  elif x == 'Excellent':
    return 4
```

```python
df['Rating Description Numerical'] = df['Rating Description'].apply(lambda x: rating_desc_to_num(x))
fig_x_y = df.groupby('Rating Description Numerical')['Rating Description Numerical'].count()
sns.barplot(x = fig_x_y.index, y = fig_x_y.values)
```
